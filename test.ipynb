```python
from docx import Document
import os
```

```python
file = os.path.join("doc","Dokumen A.docx")
doc = Document(file)
para = doc.paragraphs
for p in para:
    print(p.text)
```

```python
def printing_effect(self):
    docA = Document(self.doc1)
    docB = Document(self.doc2)

    total_diff = 0
    summary_A = {"bold": 0, "italic": 0, "underline": 0}
    summary_B = {"bold": 0, "italic": 0, "underline": 0}

    for i in range(min(len(docA.paragraphs), len(docB.paragraphs))):
        runsA = docA.paragraphs[i].runs
        runsB = docB.paragraphs[i].runs

        for j in range(min(len(runsA), len(runsB))):
            rA = runsA[j]
            rB = runsB[j]

            # hanya bandingkan jika teks sama dan tidak kosong
            if rA.text.strip() and rA.text == rB.text:

                # efek dokumen A
                effectsA = {
                    "bold": bool(rA.bold),
                    "italic": bool(rA.italic),
                    "underline": bool(rA.underline)
                }

                # efek dokumen B
                effectsB = {
                    "bold": bool(rB.bold),
                    "italic": bool(rB.italic),
                    "underline": bool(rB.underline)
                }

                # Hitung summary
                for ef in effectsA:
                    if effectsA[ef]:
                        summary_A[ef] += 1
                    if effectsB[ef]:
                        summary_B[ef] += 1

                diff_effects = 0

                for ef in effectsA:
                    if effectsA[ef] != effectsB[ef]:
                        diff_effects += 1

                if diff_effects > 0:
                    total_diff += diff_effects

                    effectA_text = ", ".join(
                        [k for k, v in effectsA.items() if v]
                    ) or "regular"

                    effectB_text = ", ".join(
                        [k for k, v in effectsB.items() if v]
                    ) or "regular"

                    print(
                        f"Dokumen A = {rA.text} - {effectA_text}\n"
                        f"Dokumen B = {rB.text} - {effectB_text}\n"
                        f"Perbedaan = {diff_effects} efek\n"
                    )

    print("\n" + "="*50)
    print("RINGKASAN EFEK DOKUMEN A:")
    print(f"  Bold: {summary_A['bold']} | Italic: {summary_A['italic']} | Underline: {summary_A['underline']}")
    print("\nRINGKASAN EFEK DOKUMEN B:")
    print(f"  Bold: {summary_B['bold']} | Italic: {summary_B['italic']} | Underline: {summary_B['underline']}")
    print(f"\nTOTAL PERBEDAAN EFEK: {total_diff}")
    print("="*50)
```

```python
printing_effect(doc)
```
